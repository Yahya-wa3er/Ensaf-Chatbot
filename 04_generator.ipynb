```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import json

print("Imports OK")
```

```python
print("Chargement du modèle... (quelques minutes la première fois)")

# On utilise un modèle GPT-2 entraîné en français, léger et efficace sur CPU
generator = pipeline(
    "text-generation",
    model="asi/gpt-fr-cased-small",
    tokenizer="asi/gpt-fr-cased-small"
)

print("Modèle chargé !")
```

```python
def generate_response(query, context, max_length=200):
    prompt = f"""Tu es l'assistant intelligent de l'ENSA Fès. Réponds à la question de l'étudiant en te basant sur le contexte fourni. Sois précis, clair et bienveillant.

Contexte : {context}

Question : {query}

Réponse :"""

    output = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=1,
        temperature=0.7,
        do_sample=True,
        pad_token_id=50256
    )
    
    # Extraire uniquement la réponse générée
    texte_complet = output[0]["generated_text"]
    reponse = texte_complet.split("Réponse :")[-1].strip()
    
    return reponse
```

```python
# Test simple
contexte = "L'ENSA Fès propose plusieurs filières d'ingénierie : Génie Informatique (GI), Génie Civil (GC), Génie Électrique (GE), Génie Mécanique (GM), et Génie Industriel (GIN). La formation dure 5 ans."

question = "Quelles filières existe à l'ENSA Fès ?"

print("Question :", question)
print("\nGénération en cours...")
reponse = generate_response(question, contexte)
print("Réponse générée :\n", reponse)
```

```python
contexte2 = "L'admission à l'ENSA Fès se fait via le concours national commun (CNC) pour les bacheliers scientifiques avec mention bien ou très bien en SM ou SP."

question2 = "Comment on peut s'inscrire à l'ENSA ?"

print("Question :", question2)
print("\nGénération en cours...")
reponse2 = generate_response(question2, contexte2)
print("Réponse générée :\n", reponse2)
```

```python
# On va créer un fichier generator.py réutilisable
code = '''
from transformers import pipeline

print("Chargement du generator...")
generator = pipeline(
    "text-generation",
    model="asi/gpt-fr-cased-small",
    tokenizer="asi/gpt-fr-cased-small"
)

def generate_response(query, context, max_length=200):
    prompt = f"""Tu es l assistant intelligent de l ENSA Fes. Reponds a la question de l etudiant en te basant sur le contexte fourni.

Contexte : {context}

Question : {query}

Reponse :"""

    output = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=1,
        temperature=0.7,
        do_sample=True,
        pad_token_id=50256
    )
    texte_complet = output[0]["generated_text"]
    reponse = texte_complet.split("Reponse :")[-1].strip()
    return reponse
'''

with open("generator.py", "w", encoding="utf-8") as f:
    f.write(code)

print("generator.py sauvegardé !")
```
